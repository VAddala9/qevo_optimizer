```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure()
df = pd.read_csv("data/example_output.csv")

g = sns.FacetGrid(data=df, col="p2", hue="optimized_for", sharey=False)
g.map(sns.lineplot, "f_in", "logical_qubit_fidelity", label="F_L")

g.set_xlabels("$F_{in}$")
g.set_ylabels("$F_{L}$")
plt.show()
```
